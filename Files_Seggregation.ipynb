```python
import os
import shutil
import pytesseract
import pdfplumber
from pdf2image import convert_from_path

# Define paths for Tesseract OCR and Poppler
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
poppler_path = r"C:\Poppler\Release-24.08.0-0\poppler-24.08.0\Library\bin"

# Main folder path
main_folder_path = r"D:\Files_Extraction_py\PDF\PHI_Attachments"
# Folder to move matched PDFs
output_folder_path = r"D:\Files_Extraction_py\PDF\Matched_PDFs"

# Text to search for
search_texts = [
    "Genetic Risk Evaluation And Testing Program",
    "Hereditary Cancer Risk Assessment"
]

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)

# Function to extract text from image-based PDF using OCR
def extract_text_from_image_pdf(pdf_path):
    images = convert_from_path(pdf_path, poppler_path=poppler_path)
    full_text = ""
    for image in images:
        text = pytesseract.image_to_string(image)
        full_text += text
    return full_text

# Function to extract text from text-based PDF
def extract_text_from_text_pdf(pdf_path):
    full_text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            if text:
                full_text += text
    return full_text

# Function to check if the PDF contains the search text
def contains_search_text(pdf_path):
    try:
        # First, try to extract text assuming it's a text-based PDF
        text = extract_text_from_text_pdf(pdf_path)
        if not text.strip():
            # If no text is found, assume it's an image-based PDF
            text = extract_text_from_image_pdf(pdf_path)
        # Check if any of the search terms are in the extracted text
        return any(search_text in text for search_text in search_texts)
    except Exception as e:
        print(f"Error processing {pdf_path}: {e}")
        return False

# Traverse through the folder and its subfolders
for root, dirs, files in os.walk(main_folder_path):
    for file in files:
        if file.endswith(".pdf"):
            pdf_path = os.path.join(root, file)
            if contains_search_text(pdf_path):
                # Move the file to the output folder
                shutil.move(pdf_path, os.path.join(output_folder_path, file))
                print(f"Moved: {pdf_path}")

print("Process completed.")
```

```python
import os
import shutil
import pytesseract
import pdfplumber
from pdf2image import convert_from_path
from fuzzywuzzy import fuzz  # For string similarity comparison

# Define paths for Tesseract OCR and Poppler
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
poppler_path = r"C:\Poppler\Release-24.08.0-0\poppler-24.08.0\Library\bin"

# Main folder path
main_folder_path = r"D:\Files_Extraction_py\P_Test_Files1\Attachments"
# Base output folder path for segregated PDFs
base_output_folder_path = r"D:\Files_Extraction_py\P_Test_Files1\Segregated_PDFs1"
# New folder to move empty subfolders
empty_folder_path = r"D:\Files_Extraction_py\P_Test_Files1\Empty_PHI_folder1"

# Define header templates for different categories
header_templates = {
    "Genetic Risk Evaluation And Testing Program": "Genetic_ass_Matched_PDFs",
    "Hereditary Cancer Risk Assessment": "Genetic_ass_Matched_PDFs",
    "TRINITY MOTHER FRANCES": "Genetic_ass_Matched_PDFs",
    "Cancer Risk Assessment Survey": "Genetic_ass_Matched_PDFs",
    "Survey Summary": "Genetic_ass_Matched_PDFs",
    "ClearPoint": "ClearPoint_Matched_PDFs",
    "med fusion": "med fusion_Matched_PDFs",
    "Clarient Diagnostic Services": "Clarient_Matched_PDFs",
    "NeoGenomics": "NeoGenomics_Matched_PDFs",
    "Lab In Hlseven": "Lab_In_Hlseven_Matched_PDFs",
    "myPlan": "myPlan_Matched_PDFs",
    "MYRIAD":"myPlan_Matched_PDFs",
    "UNIVERSITY of FLORIDA":"FLORIDA_Matched_PDFs",
    "Integrated ONCOLOGY":"Integrated_Matched_PDFs",
    "The University of Texas MD Anderson Cancer Center":"Anderson_Matched_PDFs",
    "DAVID'S MEDICAL CENTER":"DAVID_Matched_PDFs",
    "genzyme":"genzyme_Matched_PDFs",
    "Informed Consent for Panel Genetic Testing": "Genetic_ass_Matched_PDFs",
    "Patient consents to genetic risk assessment appointment":"Genetic_ass_Matched_PDFs",
    "NeoGenomics": "NeoGenomics_Matched_PDFs",
    "Did you know that cancer can run in a family":"Genetic_ass_Matched_PDFs",
    "Clarient Diagnostic Services": "Clarient_Matched_PDFs",
    "ClearPoint": "ClearPoint_Matched_PDFs",
    "MAYO CLINIC": "Mayo_Matched_PDFs",
    "GUARDANT HEALTH": "Guardant_Matched_PDFs",
    "RESPONSE GENETICS": "Response_Matched_PDFs",
    "ClearPoint": "ClearPoint_Matched_PDFs",
    "oncotype": "oncotype_Matched_PDFs",
    "CLINICAL PATHOLOGY LABORATORIES": "CLINICAL_Matched_PDFs",
    "yriad genetics":"Pierian_Matched_PDFs",
    "Pierian":"Pierian_Matched_PDFs",
    "VERISTRAT TEST RESULT REPORT":"VERISTRAT_Matched_PDFs",
    "Longview Regional":"Longview_Matched_PDFs",
    "SCOTT AND WHITE MEMORIAL HOSPITAL":"Scott_Matched_PDFs",
    "biodesix":"biodesix_Matched_PDFs",
    "ProPath":"ProPath_Matched_PDFs",
    "INVITAE":"INVITAE_Matched_PDFs",
    "THERANOSTICS":"Thernostics_Matched_PDFs",
    "Esoterix Oncology":"Esoterix_Matched_PDFs",
    "Seton Medical Center":"Seton_Matched_PDFs",
    "GENOPTIX":"GENOPTIX_Matched_PDFs",
    "AmbryGenetics":"Ambrey_Matched_PDFs"
    # Folder 3 Header
    # Add more templates as needed
}

# Define minimum similarity threshold
similarity_threshold = 92  # 92% similarity to be considered a match

# Create output folders for each template if they don't exist
for folder_name in header_templates.keys():
    output_folder = os.path.join(base_output_folder_path, folder_name)
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

# Create empty folder path if it doesn't exist
if not os.path.exists(empty_folder_path):
    os.makedirs(empty_folder_path)

# Function to extract text from image-based PDF using OCR
def extract_text_from_image_pdf(pdf_path):
    images = convert_from_path(pdf_path, poppler_path=poppler_path)
    full_text = ""
    for image in images:
        text = pytesseract.image_to_string(image)
        full_text += text
    return full_text

# Function to extract text from text-based PDF
def extract_text_from_text_pdf(pdf_path):
    full_text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            if text:
                full_text += text
                break  # We only need to extract the first page for headers
    return full_text

# Function to calculate similarity between the header and templates
def find_header_match(text, header_templates):
    for template_name, template_text in header_templates.items():
        similarity = fuzz.partial_ratio(template_text.lower(), text.lower())
        if similarity >= similarity_threshold:
            return template_name  # Return the name of the matched template
    return None

# Function to extract header and move based on similarity match
def find_matching_header_and_move(pdf_path):
    try:
        # First, try to extract text assuming it's a text-based PDF
        text = extract_text_from_text_pdf(pdf_path)
        if not text.strip():
            # If no text is found, assume it's an image-based PDF
            text = extract_text_from_image_pdf(pdf_path)

        # Extract the header and match it with the predefined templates
        matched_template = find_header_match(text, header_templates)
        if matched_template:
            # Move the file to the corresponding folder based on the matched template
            output_folder = os.path.join(base_output_folder_path, matched_template)
            shutil.move(pdf_path, os.path.join(output_folder, os.path.basename(pdf_path)))
            print(f"Moved: {pdf_path} to {output_folder}")
            return True  # Stop after the first match
        return False
    except Exception as e:
        print(f"Error processing {pdf_path}: {e}")
        return False

# Function to check and move empty subfolders
def move_empty_subfolders(root):
    # Check if subfolders in the current root are empty
    for subfolder in os.listdir(root):
        subfolder_path = os.path.join(root, subfolder)
        if os.path.isdir(subfolder_path):
            # If the folder is empty, move it
            if not os.listdir(subfolder_path):  # Check if folder is empty
                shutil.move(subfolder_path, os.path.join(empty_folder_path, subfolder))
                print(f"Moved empty folder: {subfolder_path} to {empty_folder_path}")

# Traverse through the folder and its subfolders
for root, dirs, files in os.walk(main_folder_path):
    for file in files:
        if file.endswith(".pdf"):
            pdf_path = os.path.join(root, file)
            find_matching_header_and_move(pdf_path)
    
    # After processing files in the current directory, check for empty subfolders
    move_empty_subfolders(root)

print("Process completed.")
```
